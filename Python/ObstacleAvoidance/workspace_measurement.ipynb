```python
# import and method def
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageFilter, ImageDraw
from matplotlib import cm
import time
from glob import glob

def printBeforeAndAfter(subtitle0, subtitle1, subtitle2, subtitle3, img0, img1, img2, img3):
    fig = plt.figure(figsize=(30,25))
    rows = 1
    columns = 4
    fig.add_subplot(rows,columns, 1)
    plt.title(subtitle0)
    plt.imshow(img0)
    fig.add_subplot(rows,columns, 2)
    plt.title(subtitle1)
    plt.imshow(img1)
    fig.add_subplot(rows,columns, 3)
    plt.title(subtitle2)
    plt.imshow(img2)
    fig.add_subplot(rows,columns, 4)
    plt.title(subtitle3)
    plt.imshow(img3)

def lineLength(img, left, right):
    cv2.line(img, left, right, (255,0,0), 1)
    width = right[0] - left[0]
    return width
```
